#### Query

```python
#Join JHU table with Census table, looking at relationship between senior percentage and Covid cases in each state.
```

```python
%%bigquery
SELECT GeoName, Confirmed, Recovered, Active, Estimate as Senior_percentage, Mortality_Rate
FROM JHU_staging.Nov2
JOIN Census_staging.SeniorGen
ON Province_State = GeoName
WHERE Estimate >= 20
GROUP BY GeoName, Confirmed, Recovered, Active, Estimate, Mortality_Rate
HAVING Estimate > 0
ORDER BY Estimate DESC
limit 10
```

```python
#Looking at the biggest change in difference between confirmed covid cases from Oct26 to Nov2
```

```python

%%bigquery
SELECT Nov2.Province_State, Nov2.Confirmed as Confirmed_Nov2, Oct26.Confirmed as Confirmed_Oct26, (Nov2.Confirmed - Oct26.Confirmed) as Difference
FROM JHU_staging.Nov2
JOIN JHU_staging.Oct26
ON Nov2.Province_State = Oct26.Province_State
WHERE Nov2.Confirmed > Oct26.Confirmed
GROUP By Nov2.Province_State, Nov2.Confirmed, Oct26.Confirmed
HAVING Nov2.Confirmed >0 
ORDER BY Difference DESC
limit 10
```

```python
# looking at the biggest change in active covid cases for each state between Oct12 and Nov2
```

```python
%%bigquery
SELECT Nov2.Province_State, Nov2.Active as Active_Nov2, Oct12.Active as Active_Oct12, (Nov2.Active - Oct12.Active) as Difference
FROM JHU_staging.Nov2
JOIN JHU_staging.Oct12
ON Nov2.Province_State = Oct12.Province_State
WHERE Nov2.Active > Oct12.Active
GROUP By Nov2.Province_State, Nov2.Active, Oct12.Active
HAVING Nov2.Active >10000 
ORDER BY Difference DESC
limit 10
```

```python
# looking at the biggest improvement in testing rate for covid cases for each state between Oct12 and Nov2
```

```python
%%bigquery
SELECT Nov2.Province_State, Nov2.Testing_Rate as Testing_Rate_Nov2, Oct12.Testing_Rate as Testing_Rate_Oct12, (Nov2.Testing_Rate - Oct12.Testing_Rate) as Difference
FROM JHU_staging.Nov2
JOIN JHU_staging.Oct12
ON Nov2.Province_State = Oct12.Province_State
WHERE Nov2.Testing_Rate > Oct12.Testing_Rate
GROUP By Nov2.Province_State, Nov2.Testing_Rate, Oct12.Testing_Rate
HAVING Nov2.Testing_Rate >50000 
ORDER BY Difference DESC
limit 10
```

```python
# looking at what state had the highest decrease in mortality rate between Nov2 and Oct12
```

```python
%%bigquery
SELECT Nov2.Province_State, Nov2.Mortality_Rate as Mortality_Rate_Nov2, Oct12.Mortality_Rate as Mortality_Rate_Oct12, (Nov2.Mortality_Rate - Oct12.Mortality_Rate) as Difference
FROM JHU_staging.Nov2
JOIN JHU_staging.Oct12
ON Nov2.Province_State = Oct12.Province_State
WHERE Nov2.Mortality_Rate < Oct12.Mortality_Rate
GROUP By Nov2.Province_State, Nov2.Mortality_Rate, Oct12.Mortality_Rate
HAVING Nov2.Mortality_Rate >= 0
ORDER BY Difference 
limit 10
```

```python
# looking at the percent of health workers and percent of people with insurance in the state of Texas
```

```python
%%bigquery
SELECT HealthWork.GeoID, HealthWork.GeoName, HealthWork.Estimate as PercentOfHealthWorker, Insurance.Estimate as PercentOfInsurance
FROM Census_staging.HealthWork
JOIN Census_staging.Insurance
ON HealthWork.GeoID = Insurance.GeoID
WHERE HealthWork.GeoName LIKE '%Texas%'
GROUP By HealthWork.GeoID, HealthWork.GeoName, HealthWork.Estimate, Insurance.Estimate
HAVING Insurance.Estimate >= 0
ORDER BY PercentOfHealthWorker DESC
limit 10
```

```python
# looking at Florida's highest percent of poverty and seeing the relationship to percent of seniors
```

```python
%%bigquery
SELECT PovertyGen.GeoID, PovertyGen.GeoName, PovertyGen.Estimate as PercentOfPoverty, PovSenior.Estimate as PercentOfPovSenior
FROM Census_staging.PovertyGen
JOIN Census_staging.PovSenior
ON PovertyGen.GeoID = PovSenior.GeoID
WHERE PovertyGen.GeoName LIKE '%Florida%'
GROUP By PovertyGen.GeoID, PovertyGen.GeoName, PovertyGen.Estimate, PovSenior.Estimate
HAVING PovertyGen.Estimate >= 0
ORDER BY PercentOfPoverty DESC
limit 10
```

```python
# looking at areas with the highest percent of poverty that is more than the percent of senior. Also seeing the difference between the two
```

```python
%%bigquery
SELECT PovertyGen.GeoID, PovertyGen.GeoName, PovertyGen.Estimate as PercentOfPoverty, PovSenior.Estimate as PercentOfPovSenior, (PovSenior.Estimate - PovertyGen.Estimate) as Difference
FROM Census_staging.PovertyGen
JOIN Census_staging.PovSenior
ON PovertyGen.GeoID = PovSenior.GeoID
WHERE PovertyGen.Estimate > PovSenior.Estimate
GROUP By PovertyGen.GeoID, PovertyGen.GeoName, PovertyGen.Estimate, PovSenior.Estimate
HAVING PovertyGen.Estimate >= 50
ORDER BY PercentOfPoverty DESC
limit 10
```

```python
# looking at areas with the highest percent of seniors and its relationship to percent of Seniors living in poverty
```

```python
%%bigquery
SELECT SeniorGen.GeoID, SeniorGen.GeoName, SeniorGen.Estimate as PercentOfSenior, PovSenior.Estimate as PercentOfPovSenior
FROM Census_staging.SeniorGen
JOIN Census_staging.PovSenior
ON SeniorGen.GeoID = PovSenior.GeoID
WHERE SeniorGen.Estimate > 50
GROUP By SeniorGen.GeoID, SeniorGen.GeoName, SeniorGen.Estimate, PovSenior.Estimate
HAVING PovSenior.Estimate >= 0
ORDER BY SeniorGen.Estimate DESC
```

```python
# looking at areas with the lowest percent of seniors and its relationship to the percent of people with insurance
```

```python
%%bigquery
SELECT SeniorGen.GeoID, SeniorGen.GeoName, SeniorGen.Estimate as PercentOfSenior, Insurance.Estimate as PercentOfInsurance
FROM Census_staging.SeniorGen
JOIN Census_staging.Insurance
ON SeniorGen.GeoID = Insurance.GeoID
WHERE SeniorGen.Estimate <= 10
GROUP By SeniorGen.GeoID, SeniorGen.GeoName, SeniorGen.Estimate, Insurance.Estimate
HAVING Insurance.Estimate >= 35
ORDER BY Insurance.Estimate DESC
```

```python
# join JHU with Census to look at the relationship between percent of poverty and amount of covid cases in each state 
# where percent of poverty is greater than 10
```

```python
%%bigquery
SELECT GeoName, Confirmed, Recovered, Active, PovertyGen.Estimate as PercentOfPoverty, Mortality_Rate
FROM JHU_staging.Nov2
JOIN Census_staging.PovertyGen
ON Province_State = GeoName
WHERE Estimate >= 10
GROUP BY GeoName, Confirmed, Recovered, Active, Estimate, Mortality_Rate
HAVING Mortality_Rate > 1
ORDER BY Confirmed DESC
limit 10
```
